
# Download Files from Google Drive and Setup Directory


This script downloads specific files from Google Drive using their file IDs and saves them in the `/content/` directory. It then creates a `results`
directory to store output files.

### Steps:
1. Define the file IDs and their corresponding original names.
2. Use `gdown` to download each file from Google Drive.
3. Create a directory `/content/results` if it does not already exist.




```python
import gdown
import os

# List of file IDs and their original names
files = {
    "18_Zl5Zkje4kIk7KwddYu04bY4zDPl12G": "virtenvFNM4COLAB.py",
    "1WN52O9mxx0qowMMC8wgKTwxVX8ywm8CK": "data_attack_result_FMN.pkl",
    "1EdxnaSay0BLe5JJ553L3lrZP1Jryya94": "data_attack_result_FMN_CONFIDENCE.pkl",
    "1fBtfw-USkZBMV67K6nYfw2idtizJYBmP": "data_attack_result_AA.pkl"
}

# Destination folder
output_dir = "/content/"

# Download all files with their original names
for file_id, file_name in files.items():
    url = f"https://drive.google.com/uc?id={file_id}"
    output_path = os.path.join(output_dir, file_name)
    gdown.download(url, output_path, quiet=False)

print("Download completed!")

import os

# Specify the folder path
folder_path = "/content/results"

# Create the folder if it doesn't already exist
os.makedirs(folder_path, exist_ok=True)

print(f"Folder '{folder_path}' created successfully!")
```
